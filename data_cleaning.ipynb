```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy 
import matplotlib.pyplot as plt 
```

```python
md = pd.read_csv('archive/movies_metadata.csv')
md.head().transpose()
```

```python
md.columns
```

```python
md.shape
```

```python
md.info()
```

```python
md = md.drop(['imdb_id'], axis=1)
md = md.drop(['homepage'], axis=1)
md = md.drop(['tagline'], axis=1)
```

```python
md[md['adult'] == 'False'].shape
```

We see that 45454 movies have False in the adult column (so only 12 movies are adult movies), which means we can safely discard this column as well.

```python
md = md.drop(['adult'], axis=1)
```

```python
md[md['status'] == 'Released'].shape
```

```python
md = md.drop(['status'], axis=1)
```

```python
md = md.drop(['original_title'], axis=1)
md = md.drop(['video'], axis=1)
```

```python
md[md['revenue'] == 0].shape
```

```python
md['revenue'] = md['revenue'].replace(0, np.nan)
md['original_language'] = md['original_language'].replace('xx', np.nan)
```

Now, we also convert the belongs_to_collection column to a binary of 1 or 0; 1 if the movie does belong to a collection, 0 if it does not.

```python
md['belongs_to_collection'] = md['belongs_to_collection'].fillna('').apply(lambda x: 0 if x=='' else 1)
```

```python
md['budget'] = pd.to_numeric(md['budget'], errors='coerce')
md['budget'] = md['budget'].replace(0, np.nan)
```

```python
md = md.drop(['poster_path'], axis=1)
```

Now, we remove all rows with a NaN value, leaving only the movies that have complete data.

```python
md = md.dropna()
```

```python
md.to_csv('clean_metadata.csv')
```
